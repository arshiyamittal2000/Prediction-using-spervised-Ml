```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url='https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
s_data=pd.read_csv(url)
s_data.head(10)
```

```python
s_data.plot(x='Hours',y='Scores', style= 'o')
plt.title('Hours_Study v/s Percentage_Scores')
plt.xlabel('Hours_Study')
plt.ylabel('Percentage_Scores')
plt.show()
```

```python
#Divide the data into attributes and labels.
```

```python
X=s_data.iloc[:, :-1].values
y=s_data.iloc[:, 1].values
```

```python
#Split the data inot training and tests sets. We will be using scikit learns built in train_test_split() method
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Regressor1=LinearRegression()
Regressor1.fit(X,y)
print("Training_model")
```

```python
line=Regressor1.coef_*X+Regressor1.intercept_
#plotting regression line
```

```python
plt.scatter(X,y)
plt.plot(X,line);
plt.show()
```

```python
print(X_test)
#testing data
y_pred=Regressor1.predict(X_test)

```

```python
#comparing actual vs predicted
df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```

```python
print("Training Score",Regressor1.score(X_train,y_train))
print("Testing Score",Regressor1.score(X_test,y_test))
```

```python
#plotting bar chart for actual vs predicted
df.plot(kind = 'bar',figsize = (7,7))
plt.show()
```

```python
#prediciton for 9.25 hours
hours=9.25
test=np.array([hours])
test=test.reshape(-1,1)
own_pred=Regressor1.predict(test)
print("number of hourse={}",format (hours))
print("Predicted Score= {}".format(own_pred[0]))
```

```python
import numpy as np
from sklearn  import metrics
print("Mean absolute error",metrics.mean_absolute_error(y_test,y_pred))
print("Mean squared error",metrics.mean_squared_error(y_test,y_pred))
print("Root Mean squared error",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("Explained variance score",metrics.explained_variance_score(y_test,y_pred))

#This was done to evaluate performance of algorithm
```
